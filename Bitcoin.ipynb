```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
bitcoin = pd.read_csv('source/Bitcoin.csv')
```

```python
bitcoin.head()
```

```python
bitcoin[bitcoin['High'] == bitcoin['High'].max()]
```

```python
bitcoin = bitcoin[(bitcoin.Date > '2021-04-30') & (bitcoin.Date < '2021-05-10')]
bitcoin
```

```python
bitcoin.sort_index(ascending=False, inplace=True)
bitcoin
```

```python
bitcoin.Open / bitcoin.Open.iloc[0] * 100
```

```python
d = bitcoin.Date 
o = bitcoin.Open
h = bitcoin.High
l = bitcoin.Low


plt.plot(d, o)
plt.plot(d, h)

plt.xlabel('Date')
plt.ylabel('Open and High')


plt.legend(['Open', 'High'])
plt.xticks(rotation = 90)

plt.show()
```
