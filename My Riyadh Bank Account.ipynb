```python
step1 = 1
step2 = 0
for index, row in df.iterrows():
    if (index+step1 > len(df)) | (index+step2 > len(df)):
        break
    print(f"[{index+step1}, 4] - [{index+step2}, 0]")
    df.iloc[index+step1, 4] = df.iloc[index+step2, 0]
    step1 += 1
    step2 += 1
    #print(index+1, row['التاريخ'])
```
