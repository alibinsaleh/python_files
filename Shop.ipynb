```python
import pandas as pd
from datetime import date
```

```python
df = pd.read_csv('source/items_data.csv')
```

```python
class Item:
    def __init__(self, item_id, item_name, item_description,item_quantity=0, item_wholesale_price=0, item_retail_price=0, item_discount=0):
        self.item_id = item_id
        self.item_name = item_name
        self.item_description = item_description
        self.item_quantity = item_quantity
        self.item_wholesale_price = item_wholesale_price
        self.item_retail_price = item_retail_price
        self.item_discount = item_discount
        
        
        
    def decrease_quantity(self):
        self.item_quantity -= 1
        
    def display_quantity(self):
        print(f'Quantity of: {self.item_name} is {self.item_quantity}')
```

```python
class Client:
    def __init__(self, client_id , client_firstname, client_lastname, date_created, purchases_amount=0, discounts=0):
        self.client_id = client_id
        self.client_firstname = client_firstname
        self.client_lastname = client_lastname
        self.date_created = date_created
        self.purchases_amount = purchases_amount
        self.discounts = discounts
        
        
        
    
```

```python
class Sale:
    def __init__(self, client):
        self.client = client
        self.items_purchased = []
        
    def add_item(self, item, qty):
        for i in range(qty):
            self.items_purchased.append(item)
            item.decrease_quantity()
        
    def total(self):
        sum = 0
        for item in self.items_purchased:
            sum += item.item_retail_price
            #print(item.item_name, item.item_retail_price)
        return sum 
    
    def print_invoice(self):
        print(f'Customer ID: {self.client.client_id}')
        print(f'Customer Name: {self.client.client_lastname}, {self.client.client_firstname}')
        print(f'Date: {date.today()}')
        print(''.ljust(75, '-'))
        print('#'.ljust(5),'Item'.ljust(30),'Qty'.ljust(10),'Unit Price'.ljust(15), 'Total'.ljust(10))
        print(''.ljust(75, '-'))
        for i, item in enumerate(self.items_purchased):
            tot = item.item_quantity * item.item_retail_price
            print(str(i).ljust(5),item.item_name.ljust(30),str(item.item_quantity).ljust(10),str(item.item_retail_price).ljust(15), str(tot).ljust(10))
```

```python
ali = Client(101, 'Ali', 'Almohammed Saleh', date.today())
item1 = Item(1001, 'Coca Cola', 'Coca Cola', 100, 2.5, 3)
# item1.add_to_dataframe()
item2 = Item(1002, 'Napkin', 'Napkin', 100, 20, 24)
# item2.add_to_dataframe()
item3 = Item(1003, 'Chedar Cheese', 'Chedar Cheese', 50, 15, 18)
# item3.add_to_dataframe()
item4 = Item(1004, 'Milk', 'Milk', 150, 9, 11)
# item4.add_to_dataframe()
item5 = Item(1005, 'Noga Buiskets', 'Noga Buiskets', 40, 24, 26)
# item5.add_to_dataframe()
```

```python
sale1 = Sale(ali)
sale1.add_item(item1, 13)
sale1.add_item(item2, 1)
sale1.add_item(item3, 5)
sale1.add_item(item4, 2)
sale1.add_item(item5, 30)
```
