```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv('source/digital_tech_1_1.csv')
```

```python
df
```

### Top 10 students 

```python
top_20 = df.sort_values(by='total', ascending=False).head(20)
top_20[['classroom', 'student_name', 'total']]
```

```python
sns.distplot(top_20['total'], bins=5)
```

### Bottom 5 students

```python
bottom_20 = df.sort_values(by='total', ascending=True).head(20)
bottom_20[['classroom', 'student_name', 'total']]
```

```python
sns.countplot(x='total', data=top_20)
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
plt.title('Top 20 Students')
sns.barplot(x='student_english_name', y='total', data=top_20, palette='Set2')
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
plt.title('Bottom 20 Students')
sns.barplot(x='student_english_name', y='total', data=bottom_20, palette='Set2')
```

### Highest overall performance classrooms

```python
# Top countries participating

top_classroom = top_20.classroom.value_counts().sort_values(ascending=False).head(3)
top_classroom
```

```python
# Plot for the top 10 countries

#plt.figure(figsize=(12, 6))
#plt.xticks(rotation=20)
plt.title('                   ')
sns.barplot(x=top_classroom.index, y=top_classroom, palette='Set2')
```
