### Table Printer

```python
tableData = [
                ['apples', 'oranges', 'cherries', 'banana'],
                ['Alice', 'Bob', 'Carol', 'David'],
                ['dogs', 'cats', 'moose', 'goose']
            ]

def findLongest(l):
    longest = len(tableData[0][0])
    for i in range(4):
        for j in range(len(tableData)):
            if len(tableData[j][i]) > longest:
                longest = len(tableData[j][i])
            print(len(tableData[j][i]), end=" ")
        
        print('\n')
        
    return longest

indent = findLongest(tableData)

for i in range(4):
    for j in range(len(tableData)):
        print((tableData[j][i]).rjust(indent), end=" ")
    print("\n")
```

## Search for word or words in a clipboard text

```python
import pyperclip

text = pyperclip.paste()
word = 'lorem'
count = 0

for i in range(len(text)):
    #print(i, text[i:len(word)+i])
    if text[i:len(word)+i].upper() == word.upper():
        print(text[i:len(word)+i])
        count += 1
print(f'Number of occurance for the word {word} is {count}')
```
